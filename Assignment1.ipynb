```python
import nltk
import string
from nltk.tokenize import (
    WhitespaceTokenizer,
    TreebankWordTokenizer,
    TweetTokenizer,
    MWETokenizer,
    word_tokenize
)
from nltk.stem import PorterStemmer, SnowballStemmer, WordNetLemmatizer

nltk.download('punkt')
nltk.download('wordnet')
```

```python
text = "Natural Language Processing is amazing! Let's analyze text properly."
```

Whitespace Tokenizer

```python
WhitespaceTokenizer().tokenize(text)
```

```python
clean_text = text.translate(str.maketrans('', '', string.punctuation))
word_tokenize(clean_text)
```

TreeBankWord Tokernizer

```python
TreebankWordTokenizer().tokenize(text)
```

Tweet Tokernizer

```python
TweetTokenizer().tokenize(text)
```

Multiword Expression Tokernizer

```python
mwe = MWETokenizer([('Natural', 'Language', 'Processing')], separator='_')
mwe.tokenize(word_tokenize(text))
```

Porter Stemmer

```python
porter = PorterStemmer()
[porter.stem(word) for word in word_tokenize(clean_text)]
```

Snowball Stemmer

```python
snowball = SnowballStemmer("english")
[snowball.stem(word) for word in word_tokenize(clean_text)]
```

Lemmatization

```python
lemmatizer = WordNetLemmatizer()
[lemmatizer.lemmatize(word.lower()) for word in word_tokenize(clean_text)]
```
