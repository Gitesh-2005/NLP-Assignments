```python
corpus = [
    "I love natural language processing",
    "NLP loves Python",
    "Python is powerful for NLP"
]
```

Bag of Words (Count Occurrence)

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
bow_sparse = cv.fit_transform(corpus)

feature_names = cv.get_feature_names_out()
bow = bow_sparse.toarray()

feature_names, bow
```

```python
from sklearn.preprocessing import normalize

# CountVectorizer doesn't support `norm`. Reuse the already-created BoW matrix and normalize it.
bow_norm = normalize(bow_sparse, norm="l1", axis=1)
bow_norm.toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(corpus)
tfidf_matrix.toarray()
```

```python
tokenized_corpus = [sentence.lower().split() for sentence in corpus]
```

```python
from gensim.models import Word2Vec

model = Word2Vec(
    sentences=tokenized_corpus,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4
)

# quick sanity check (optional)
model.wv["python"]
```
